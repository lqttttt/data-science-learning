```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# 读取数据
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")

print(test_data.head())
# 选择特征
data_features = ["SibSp", "Parch", "Fare","Age","Pclass","Embarked","Sex"]

# 处理训练数据
X = pd.get_dummies(train_data[data_features])
y = train_data["Survived"]

# 处理测试数据
XX = pd.get_dummies(test_data[data_features])

# 创建填充器（用训练数据的中位数）
imputer = SimpleImputer(strategy='median')
X_filled = imputer.fit_transform(X)
XX_filled = imputer.transform(XX)

# 分割训练验证集
train_X, val_X, train_y, val_y = train_test_split(X_filled, y, test_size=0.1,random_state=42)

# 训练模型
forest_model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
forest_model.fit(train_X, train_y)
pred_test= forest_model.predict(val_X)
accuracy=accuracy_score(val_y,pred_test)
print(accuracy)
# 预测
predictions = forest_model.predict(XX_filled)
```

```python
output=pd.DataFrame({"PassengerId":test_data.PassengerId,"Survived":predictions})
output.to_csv("submission.csv",index=False)
print("Your submission was succesfully saved")
```
